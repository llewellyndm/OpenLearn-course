Cape Town is the capital city of South Africa. It typically has a hot dry climate in the summer months (December to March) and a mild and moderately wet climate in the winter months (June to August). We assume that the reader lives in the UK and during our dreary winter season wants to go abroad somewhere warmer, such as Cape Town where it would be summer. So we will look at weather data for the months of January, February and March in Cape Town to determine which two-week period would be the best to visit.

## Getting the data
The data has been taken from an online course by the Open University entitled ['Learn to code for data analysis'](https://www.open.edu/openlearn/science-maths-technology/learn-code-data-analysis/content-section-overview-0?active-tab=description-tab). However, the original source of the data was a website called [Weather Underground](https://www.wunderground.com/). It contains information about the weather in Cape Town for every day of 2014.

```python

from pandas import *
df = read_csv('CapeTown_CPT_2014.csv', skipinitialspace=True)
df
```

Currently the 'Date' column has string entries, and we want these to be dates so that we can filter the dataframe to only include the first three months of the year.

```python
df['Date'] = to_datetime(df['Date'])
```

We now re-index the dataframe so that each row has a date as its index.

```python
df.index = df['Date']
```

Now we can filter the data so that we only have the summer months.

```python
from datetime import datetime
summer = df.loc[datetime(2014,1,1) : datetime(2014,3,31)]
summer
```

## Cleaning the data
We have a dataframe for the summer months in Cape Town now, as seen above. However, some cleaning would not go amiss. Firstly we need to sort out the final column, which is the wind direction. The column name includes '<br/>' at the end which is simply a vestige of the website's html code. The same applies for each entry in the rows. We can simply rename the column and remove the end off each number, and then change the type of data into a number.

```python
summer = summer.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
summer['WindDirDegrees'] = summer['WindDirDegrees'].str.rstrip('<br />')
```

Now we want to sort out the Max Gust Speed column. Many days have no data and so have an entry of NaN. 

```python
summer['Max Gust SpeedKm/h'] = summer['Max Gust SpeedKm/h'].fillna('')
summer
```

We need to do the same for 'Events'.

```python
summer['Events'] = summer['Events'].fillna('')
summer
```

We also need to check that all the data types are appropriate.

```python
summer.dtypes
```

As can be seen above, <code>'WindDirDegrees'</code> has the type <code>object</code>. We need to change these to whole numbers i.e. the <code>int64</code> type.

```python
summer['WindDirDegrees'] = summer['WindDirDegrees'].astype('int64')
summer.dtypes
```

### Analysing the data

Now let's look at the data to see if there are any periods with a favourable temperature.

```python
%matplotlib inline
summer['Mean TemperatureC'].plot(grid=True, figsize=(10,5))
```

The temperature ranges between 16 and 28 degrees Celsius and actually seems to vary quite a bit day-to-day. However the temperature does seem to generally be highest in February. How about rainfall?

```python
summer[['Mean TemperatureC', 'Precipitationmm']].plot(grid=True, figsize=(10,5))
```

There seems to be the least amount of rain for the period from 13th January to 1st March, and hen from 3rd March to 18th March. Seeing as we had nicer temperatures in January and February, maybe the holiday should be booked for sometime then.

## Conclusion
It appears that due to the warm but not too high temperatures found in the beginning of February, as well as the low amount of rain, that the ideal period to journey to Cape Town would be the first two weeks of February.
