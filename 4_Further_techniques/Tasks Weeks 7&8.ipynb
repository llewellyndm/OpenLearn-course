## Task 4
Using your own monthly data for a single year, which countries only trade in your selected trade item rarely or for small amounts? Which partners trade on a regular basis (for example, in at least nine of the months)?

Can you also find countries that trade regularly but only for small amounts (for example whose maximum monthly trade value is less than a certain threshold amount) or who trade infrequently but for large amounts (or other combinations thereof)?



```python
honey_countries_imports.groupby('Partner')['Trade Value (US$)'].aggregate(sum).sort_values(ascending=False).plot(logx=True, figsize=(20,10), kind='barh')
```

```python
def weakPartner(g):
    return len(g) <= 3 | g['Trade Value (US$)'].sum() < 10000 

weak_countries_imports = honey_countries_imports.groupby(['Partner']).filter(weakPartner)
weak_countries_imports.groupby(['Partner'])[['Trade Value (US$)']].aggregate([len, sum])
```

```python
def regPartners(g):
    return len(g) >= 9

regular_import_partners = honey_countries_imports.groupby('Partner').filter(regPartners)
regular_import_partners.groupby('Partner')[['Trade Value (US$)']].aggregate([len, sum]).sort_values(('Trade Value (US$)', 'sum'))
```

## Task 5

```python
def moreThan100Thousand(g):
    return g['Trade Value (US$)'].sum() > 100000

def moreThan500Thousand(g):
    return g['Trade Value (US$)'].sum() > 500000
```

```python
EUROPE = ['France', 'Spain', 'Belgium', 'Germany', 'Italy']
honey_countries_EUR = honey_countries[honey_countries['Partner'].isin(EUROPE)]

MONTHS = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
def toMonth(n):
    return MONTHS[n-201401]

honey_countries_EUR['Month'] = honey_countries_EUR['Period'].apply(toMonth)
```

```python
pivot_table(honey_countries_EUR.groupby(['Partner', 'Trade Flow','Month']).filter(moreThan500Thousand),
            #filter works by chekcing each combination of (Partner, Trade Flow, Period) to see whether that entry
            #has a value greater than 500,000
            index = ['Partner','Trade Flow'],
            columns = 'Month',
            values='Trade Value (US$)',
            aggfunc=sum)
```

```python
pivot_table(weak_countries_imports,
           index=['Partner'],
           values='Trade Value (US$)',
           columns='Period',
           aggfunc=sum)
```
